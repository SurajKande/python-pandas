```
 """Function that takes a row of data,
    drops all missing values,
    and checks if all remaining values are greater than or equal to 0
    """
def check_null_or_valid(row_data):
     
    no_na = row_data.dropna()
    numeric = pd.to_numeric(no_na)
    ge0 = numeric >= 0
    return ge0

# Check whether the values in the row are valid
assert df.iloc[:, 1:].apply(check_null_or_valid, axis=1).all().all()
```

```
#repeating the same above steps for 20th and 21st( upto 2016 ) centuries

# dataset is the life expectency of 20th century
file_name1 = '/content/gdrive/My Drive/datasets/g1900.csv'
df1 = pd.read_csv(file_name1)

# dataset is the life expectency of 21th century
file_name2 = '/content/gdrive/My Drive/datasets/g2000.csv'
df2 = pd.read_csv(file_name2)
```

```
df1.head()
```

```
df2.head()
```

```
df.set_index('Life expectancy', inplace  =True)
df1.set_index('Life expectancy', inplace  =True)
df2.set_index('Life expectancy', inplace  =True)
```

```
#concating the data frames into one dataframe
life_expectency_dataframe = pd.concat([df, df1, df2], axis=1)
#type(life_expectency_dataframe)
```

```
life_expectency_dataframe.head()
```

```
life_expectency_dataframe.reset_index(inplace=True)
```

```
life_expectency_dataframe.head()
```

```
#Reshaping the data to a proper format

#using melt to reshape the data
life_expectency_melt =  pd.melt ( life_expectency_dataframe, id_vars = 'Life expectancy', var_name='year', value_name= 'life_expectancy')
```

```
life_expectency_melt.rename(columns = {'Life expectancy':'country'}, inplace=True)
```

```
life_expectency_melt
```

```
# Convert the year column to numeric
life_expectency_melt.year = pd.to_numeric(life_expectency_melt.year)

# Test if country is of type object
assert life_expectency_melt.country.dtypes == np.object

# Test if year is of type int64
assert life_expectency_melt.year.dtypes == np.int64

# Test if life_expectancy is of type float64
assert life_expectency_melt.life_expectancy.dtypes == np.float64
```

```
# To check in the country for any wrong data
pattern = '^[A-Za-z .]*$'

mask = life_expectency_melt.country.str.contains(pattern)             
print(mask)
```

```
#there are few false i.e there are some countries with wrong input
invalid_countries = life_expectency_melt.country[~mask]               #~mask is to select the False values, basically we inverted the values to access the false values
print(invalid_countries)
```

```
# checking that country does not contain any missing values
assert pd.notnull(life_expectency_melt.country).all()

# checking that year does not contain any missing values
assert pd.notnull(life_expectency_melt.year).all()
```

```
# Group life_expectency_melt
life_expectency_agg = life_expectency_melt.groupby('year')['life_expectancy'].mean()

print(type(life_expectency_agg))
life_expectency_agg.head()
```

```
# Add first subplot
plt.subplot(2, 1, 1) 

# Create a histogram of life_expectancy
life_expectency_melt['life_expectancy'].plot(kind="hist")

# Add second subplot
plt.subplot(2, 1, 2)

# Create a line plot of life expectancy per year
life_expectency_agg.plot()

# Add title and specify axis labels
plt.title('Life expectancy over the years')
plt.ylabel('Life expectancy')
plt.xlabel('Year')

# Display the plots
plt.tight_layout()
plt.show()
```
